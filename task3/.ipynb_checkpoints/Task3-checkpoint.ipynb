```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

print(os.getcwd())
```

```python
dataset = pd.read_excel('barrettII_eyes_clustering.xlsx')
dataset.shape
dataset
```

```python
from sklearn.cluster import KMeans
```

```python
X = dataset.copy()
X = X.drop(['Correto', 'ID'], axis = 1)
X
```

```python
kmeans2 = KMeans(n_clusters=2, random_state=0).fit(X)
ykm2 = kmeans2.predict(X)
kmeans3 = KMeans(n_clusters=3, random_state=0).fit(X)
ykm3 = kmeans3.predict(X)
kmeans4 = KMeans(n_clusters=4, random_state=0).fit(X)
ykm4 = kmeans4.predict(X)
kmeans5 = KMeans(n_clusters=5, random_state=0).fit(X)
ykm5 = kmeans5.predict(X)

centers_km2 = kmeans2.cluster_centers_
centers_km3 = kmeans3.cluster_centers_
centers_km4 = kmeans4.cluster_centers_
centers_km5 = kmeans5.cluster_centers_

print("Centroids -> {} Result {}".format(centers_km2, len(ykm2)))

print()
print()
print()
```

```python
feature_columns = ['AL', 'ACD', 'WTW', 'K1', 'K2']

(unique, counts) = np.unique(ykm2, return_counts=True)
frequencies = np.asarray((unique, counts)).T
quantity = frequencies[:,1]

print("Centroids \n {} \n Quantity \n {}".format(centers_km2, frequencies))

result2 = pd.DataFrame(centers_km2, columns = feature_columns)
result2['Quantity'] = quantity
result2
```
